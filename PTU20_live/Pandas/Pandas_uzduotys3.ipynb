```python
def to_number(num):
    final = ""
    if type(num) == int:
        return num
    else:
        for x in str(num):
            if x.isdecimal() or x == ".":
                final += x
        if len(final) > 0:
            return int(float(final))
        return int(0)
    
```

```python
to_number("kok387lkdsd")
```
